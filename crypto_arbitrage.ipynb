#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = coinbase_sliced['Close'].loc['2018-01-16'] - bitstamp_sliced['Close'].loc['2018-01-16']

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
early_filter = arbitrage_spread_early > 0.0
arbitrage_spread_early.loc[early_filter].describe()
```

```python
# For the middle date, measure the arbitrage spread
arbitrage_spread_middle = coinbase_sliced['Close'].loc['2018-02-24'] - bitstamp_sliced['Close'].loc['2018-02-24']

# Use a conditional statement to generate the middle date summary statistics
middle_filter = arbitrage_spread_middle > 0.0
arbitrage_spread_middle.loc[middle_filter].describe()
```

```python
# For the late date, measure the arbitrage spread
arbitrage_spread_late = coinbase_sliced['Close'].loc['2018-03-26'] - bitstamp_sliced['Close'].loc['2018-03-26']

# Use a conditional statement to generate the late date summary statistics
late_filter = arbitrage_spread_late > 0.0
arbitrage_spread_late.loc[late_filter].describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = arbitrage_spread_early.loc[early_filter] / bitstamp_sliced['Close'].loc['2018-01-16']

# Review the spread return DataFrame
spread_return_early.head()
```

```python
# For the middle date, calculate the spread returns
# I kept getting all NaN values returned and I can't figure out why so I used dropna() to
# drop the NaN values and it fixed it. Not sure why this was stubborn with the timestamps and
# the early and late versions weren't, but the dropna() seems to fix the issue
spread_return_middle = arbitrage_spread_middle.loc[middle_filter] / bitstamp_sliced['Close'].loc['2018-02-24']
spread_return_middle = spread_return_middle.dropna()
spread_return_middle.head()
```

```python
# For the late date, calculate the spread returns
spread_return_late = arbitrage_spread_late.loc[late_filter] / bitstamp_sliced['Close'].loc['2018-03-26']

# Review the first 5 rows
spread_return_late.head()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early = spread_return_early[spread_return_early > .01]

# Review the first five profitable trades
profitable_trades_early.head()
```

```python
# Calculate the number of profitable trades for the early date
early_num = len(profitable_trades_early)
print(f"The number of profitable trades for the early date is: {early_num}")
```

```python
# Find trades from the middle date with positive returns that exceed the 1% threshold
profitable_trades_middle = spread_return_middle[spread_return_middle > .01]

# Review the profitable trades
profitable_trades_middle
```

```python
# Calculate the number of profitable trades for the middle date
middle_num = len(profitable_trades_middle)
print(f"The number of profitable trades for the middle date is: {middle_num}")
```

```python
# Find trades from the late date with positive returns that exceed the 1% threshold
profitable_trades_late = spread_return_late[spread_return_late > .01]

# Review the profitable trades
profitable_trades_late
```

```python
# Calculate the number of profitable trades from the late date
late_num = len(profitable_trades_late)
print(f"The number of profitable trades for the late date is: {late_num}")
```

## ***The average returns and the number of profitable trades both progressively decreased over the course of the three months.***

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the date early in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp_sliced['Close'].loc["2018-01-16"]

# Drop any missing values from the profit DataFrame
profit_per_trade_early = profit_early.dropna()

# View the early profit DataFrame
profit_per_trade_early
```

```python
# Calculate potential profit for the middle date
profit_middle = profitable_trades_middle * bitstamp_sliced['Close'].loc["2018-02-24"]

# Drop any missing values
profit_per_trade_middle = profit_middle.dropna()

# View the results
profit_per_trade_middle
```

#### 7. Calculate the potential arbitrage profits that you can make on each day. To do so, sum the elements in the profit_per_trade DataFrame.
